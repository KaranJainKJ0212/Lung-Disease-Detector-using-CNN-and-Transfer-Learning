```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import VGG16, VGG19, ResNet50, DenseNet121, InceptionV3, Xception
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, LearningRateScheduler
from sklearn.metrics import classification_report, confusion_matrix

# ==========================
# 🚀 DATA GENERATOR (70% Train, 15% Validation, 15% Test)
# ==========================
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.3)  # 30% for validation + test

def load_data(dataset_path, target_size=(224, 224), batch_size=64):
    train_gen = datagen.flow_from_directory(
        dataset_path,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )
    
    valid_test_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.5)  # 50% of 30% (15%)
    val_gen = valid_test_datagen.flow_from_directory(
        dataset_path,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=True
    )
    test_gen = valid_test_datagen.flow_from_directory(
        dataset_path,
        target_size=target_size,
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=False
    )
    
    return train_gen, val_gen, test_gen

# ==========================
# 🚀 MODEL ARCHITECTURE
# ==========================
def create_model(base_model_name):
    base_models = {
        'VGG16': VGG16,
        'VGG19': VGG19,
        'ResNet50': ResNet50,
        'DenseNet121': DenseNet121,
        'InceptionV3': InceptionV3,
        'Xception': Xception
    }
    
    base_model = base_models[base_model_name](weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_model.trainable = False  # Freeze base model layers
    
    inputs = tf.keras.Input(shape=(224, 224, 3))
    x = base_model(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(3, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

# ==========================
# 🚀 TRAIN AND EVALUATE MODEL (Including Full Test Set)
# ==========================
def train_and_evaluate(model, train_gen, val_gen, test_gen, base_model_name):
    model.compile(optimizer=Adam(learning_rate=3e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=30,
        callbacks=[
            EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-7)
        ],
        verbose=2
    )
    
    # ==========================
    # 🚀 FULL TEST SET EVALUATION
    # ==========================
    y_true = []
    y_pred = []
    
    for _ in range(len(test_gen)):
        test_images, test_labels = next(test_gen)
        predictions = model.predict(test_images)
        
        y_true.extend(np.argmax(test_labels, axis=1))
        y_pred.extend(np.argmax(predictions, axis=1))
    
    # Print classification report
    print(f"\nClassification Report for {base_model_name} on Full Test Set:")
    print(classification_report(y_true, y_pred, target_names=train_gen.class_indices.keys()))
    
    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=train_gen.class_indices.keys(), yticklabels=train_gen.class_indices.keys())
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title(f"Confusion Matrix - {base_model_name} (Full Test Set)")
    plt.show()

# ==========================
# 🚀 RUN EXPERIMENTS
# ==========================
dataset_path = "/kaggle/input/lung-disease-datasets/Lung X-Ray Image 2"
train_gen, val_gen, test_gen = load_data(dataset_path)

base_models = ['VGG16', 'VGG19', 'ResNet50', 'DenseNet121', 'InceptionV3', 'Xception']

for base_model_name in base_models:
    print(f"\nTraining with {base_model_name}...")
    model = create_model(base_model_name)
    train_and_evaluate(model, train_gen, val_gen, test_gen, base_model_name)
```

```python
import cv2
import numpy as np
import os
import random
import matplotlib.pyplot as plt

# Function to apply Gaussian Blur
def apply_gaussian_blur(image):
    return cv2.GaussianBlur(image, (7, 7), 0)

# Function to apply Salt & Pepper Noise
def apply_salt_pepper_noise(image, salt_prob=0.02, pepper_prob=0.02):
    noisy_image = np.copy(image)
    total_pixels = image.shape[0] * image.shape[1]
    
    # Add Salt (white pixels)
    num_salt = int(total_pixels * salt_prob)
    coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1]] = (255, 255, 255)

    # Add Pepper (black pixels)
    num_pepper = int(total_pixels * pepper_prob)
    coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape[:2]]
    noisy_image[coords[0], coords[1]] = (0, 0, 0)

    return noisy_image

# Define dataset path
dataset_path = "/kaggle/input/lung-disease-datasets/Lung X-Ray Image 2"

# Load one sample image from each class
class_folders = ["Healthy", "Lung_Abnormality", "Viral Pneumonia"]
sample_images = {}

for class_folder in class_folders:
    class_path = os.path.join(dataset_path, class_folder)
    image_files = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
    if image_files:
        sample_images[class_folder] = os.path.join(class_path, random.choice(image_files))

# Process and display images
fig, axes = plt.subplots(len(sample_images), 3, figsize=(12, 9))
for i, (label, image_path) in enumerate(sample_images.items()):
    # Load the image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply transformations
    gaussian_blurred = apply_gaussian_blur(image)
    salt_pepper_noised = apply_salt_pepper_noise(image)

    # Display results
    axes[i, 0].imshow(image)
    axes[i, 0].set_title(f"Original - {label}")

    axes[i, 1].imshow(gaussian_blurred)
    axes[i, 1].set_title("Gaussian Blur")

    axes[i, 2].imshow(salt_pepper_noised)
    axes[i, 2].set_title("Salt & Pepper Noise")

    for ax in axes[i]:
        ax.axis("off")

plt.tight_layout()
plt.show()
```
